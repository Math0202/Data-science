```python
import pandas as pd
dataset=pd.read_csv(r"C:\Users\tunam\OneDrive - NUST\Documents\Desktop\lab 1 and 2\images\metadata.csv")
dataset
```

```python
#deleting columns that are not needed
dataset.drop(['rarespecies_id','eol_content_id','eol_page_id','kingdom','sciName','common','file_name'],axis=1, inplace=True)
dataset
```

```python
#deleting duplicates directly on the original dataset without creating a new one
dataset.drop_duplicates(inplace=True)
dataset
```

```python
#deleting empty rows
dataset.dropna(inplace=True)
dataset
```

```python
#resets the index of the table to be in a numerical order 
#drops the index column with previous index value 
dataset.reset_index(drop=True, inplace=True)
dataset
```

```python
#shuffling the dataset and returning all (100%) of the rows but shuffled 
dataset= dataset.sample(frac=1)
dataset
```

```python
#renumbering the row indices
dataset.reset_index(drop=True, inplace=True)
dataset
```

Encoding

```python

from sklearn.preprocessing import LabelEncoder



# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to all columns (assuming all are categorical)
for column in dataset.columns:
    dataset[column] = label_encoder.fit_transform(dataset[column])

# Display the encoded dataset
print(dataset.head())
```

```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn import model_selection
from sklearn.model_selection import train_test_split


input_columns = ['phylum', 'genus', 'class', 'family', 'species']  # Example input columns
output_column = 'order'  # Example output column

# Define X and Y
X = dataset[input_columns]  # Input features
Y = dataset[output_column]   # Output/target variable

# Split the dataset into training and testing sets (e.g., 80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


# Check for missing values
print("Checking for missing values:")
print(X_train.isnull().sum())

# Fill missing values (if any)
X_train.fillna(X_train.mean(), inplace=True)  # This replaces NaNs with the mean of each column

# Ensure all columns in X_train are of numeric type
X_train = X_train.select_dtypes(include=[float, int])

# Check the shape of Y_train
print("Shape of Y_train:", Y_train.shape)

# Make sure Y_train is a 1D array if it's currently a DataFrame
if len(Y_train.shape) > 1:
    Y_train = Y_train.values.ravel()  # Flatten the array if necessary

# Define models
models = []
models.append(('LR', LogisticRegression(max_iter=200)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))

# Evaluate each model
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=10, random_state=42, shuffle=True)
    try:
        cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy', error_score='raise')
        results.append(cv_results)
        names.append(name)
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)
    except Exception as e:
        print(f"Error with model {name}: {str(e)}")
```
