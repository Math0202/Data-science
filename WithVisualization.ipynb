```python



phylum_counts = dataset['phylum'].value_counts().reset_index()
phylum_counts.columns = ['phylum', 'frequency']


colors = ['grey', 'blue', 'purple', 'darkgreen']


def highlight_phylum(phylum_to_highlight):

    pull_values = [0.1 if phylum == phylum_to_highlight else 0 for phylum in phylum_counts['phylum']]
    

    fig = px.pie(phylum_counts, names='phylum', values='frequency', 
                 labels={'phylum': 'The Phylum', 'frequency': 'Frequency'},
                 color_discrete_sequence=colors)

    fig.update_traces(pull=pull_values)
    fig.update_layout(height=600, width=600,
             
    title={
        'text':f"The Proportion of {phylum_to_highlight} phylum" ,
        'font': {'size': 24, 'color': 'navy', 'family': 'Arial, sans-serif', 'weight': 'bold'}
    },
    xaxis_title={
        'text': 'The Phylum',
        'font': {'size': 18, 'color': 'navy', 'family': 'Arial, sans-serif', 'weight': 'bold'}
    },
    yaxis_title={
        'text': 'Phylum_counts',
        'font': {'size': 18, 'color': 'navy', 'family': 'Arial, sans-serif', 'weight': 'bold'}
    },
    hoverlabel=dict(
        bgcolor="white",
        font_size=16,
        font_family="Rockwell"
    ))

    fig.show()
for phylum in phylum_counts['phylum']:
    highlight_phylum(phylum)
    

```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
# Apply Label Encoding to all categorical columns
label_encoder = LabelEncoder()
```

```python
 #Identifying columns that need encoding
categorical_columns = ['class', 'order', 'family', 'genus', 'species',]
```

```python
# Encoding each categorical column
for column in categorical_columns:
    dataset[column] = label_encoder.fit_transform(dataset[column])
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn import model_selection
```

```python
array = dataset.values
X = array[:, 0:5]
Y = array[:, 5]
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation =model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
seed = 42
scoring = 'accuracy'
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
# print(models)
# evaluate each model in turn
results = []
names = []
for name, model in models:
        kfold = model_selection.KFold(n_splits=10, random_state=None)
        cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
        results.append(cv_results)
        names.append(name)
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)
```
